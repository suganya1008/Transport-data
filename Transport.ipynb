```python
from scipy import stats
```

```python
def chi(d1,d2):
  for i in range (0,40):
   return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] <0.02 else False
```

```python
chi(df["destination"],df["fare"])
```
